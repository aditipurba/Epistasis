```python
def initialization(n,mac):   
    num_gene = n*mac
    population_list=[]
    for i in range(population_size):
        nxm_random_num=list(np.random.permutation(num_gene)) # generate a random permutation of 0 to num_job*num_mc-1
        population_list.append(nxm_random_num) # add to the population_list
        for j in range(num_gene):
            population_list[i][j]=population_list[i][j]%n
    return population_list



def calculate(population_list,n,mac,pt,ms):
    chrom_fitness,chrom_fit=[],[]
    total_fitness=0
    
    
    for m in range(n):
        j_keys=[j for j in range(n)]
        key_count={key:0 for key in j_keys}
        j_count={key:0 for key in j_keys}
        m_keys=[j+1 for j in range(mac)]
        m_count={key:0 for key in m_keys}
        
        for i in population_list[m]:
            gen_t=int(pt[i][key_count[i]])
            gen_m=int(ms[i][key_count[i]])
            #gen_t=int(pt[0][ms[m][i]])
            #gen_m=int(ms[m][i])
            j_count[i]=j_count[i]+gen_t
            m_count[gen_m]=m_count[gen_m]+gen_t
            
            if m_count[gen_m]<j_count[i]:
                m_count[gen_m]=j_count[i]
            elif m_count[gen_m]>j_count[i]:
                j_count[i]=m_count[gen_m]
            
            key_count[i]=key_count[i]+1
    
        makespan=max(j_count.values())
        
        #chrom_fitness.append(1/makespan)
        chrom_fit.append(makespan)
        #total_fitness=total_fitness+chrom_fitness[m]
    return chrom_fit    
```

```python
def neighborhoodINS(N,x):
    
    r1 = random.randint(0,int(N/2) - 1)
    r2 = random.randint(int(N/2) + 1, N - 1)
    for i in range(0,N):
        for j in range(0,N):
            if i!=j and i!=j+1:
                temp=list(x)
                temp.pop(i)
                temp.insert(j,i+1)
                if i==r1 and j==r2:
                    #min = dist
                    ans = temp
              
    return ans


def neighborhoodSNS(N,x):
    r1 = random.randint(0,int(N/2))
    r2 = random.randint(int(N/2) + 1, N - 1)
    for i in range(0,N):
        for j in range(i+1,N):
            temp=list(x)
            temp[i],temp[j] = temp[j],temp[i] 
            if i == r1 and j == r2:
                ans = temp 
    return ans


def neighborhoodRNS(N,x):
    r1 = random.randint(0,int(N/2))
    r2 = random.randint(int(N/2) + 1, N - 1)
    for i in range(0,N-1):
        for j in range(i+1,N):
            temp=list(x)
            p1 = j
            p2 = i
            while p1 > p2:
                temp3 = temp[p1]
                temp[p1] =temp[p2]
                temp[p2] = temp3
                p1 = p1- 1
                p2 =p2 + 1
            if i == r1 and j == r2:
                ans = temp
    return ans


def neighborhoodBNS(N,x,a):
    r1 = random.randint(0,int(N/2))
    r2 = random.randint(int(N/2) + 1, N - 1)
    for i in range(0,N-1):
        for j in range(i+1,N):
                temp=list(x)
                temp[i:j+1] = reversed(a[i:j+1])
                if i == r1 and j == r2:
                    ans = temp
    return ans
```

```python
def ranper(n):  
    pop = []
    for i in range(n):
        p = list(np.random.permutation(n)+1)
        while p in pop:
            p = list(np.random.permutation(n)+1)
        pop.append(p)
    return pop

def binarySearch(arr, l, r, x): 
  
    while l <= r: 
  
        mid = int(l + (r - l)/2 )
        if arr[mid] == x: 
            return mid 
        elif arr[mid] < x: 
            l = mid + 1 
        else: 
            r = mid - 1
    return -1

def fact(popu):
    dec=[]
    for pi in popu:
        N=0
        P=[i for i in range(1,len(pi)+1)]
        F={}
        D=[i for i in range(len(pi))]
        for i in range(len(pi)):
            index=binarySearch(P,0,len(P)-1,pi[i])
            F[i] = D[index]
            del P[index]
            N= N+ F[i]*math.factorial(D[n-i-1])
        dec.append(N)
    return dec
```

```python
pt_tmp=pd.read_excel("abzf.xlsx",sheet_name="Processing Time",index_col =[0])
ms_tmp=pd.read_excel("abzf.xlsx",sheet_name="Machines Sequence",index_col =[0])

dfshape=pt_tmp.shape
mac=dfshape[1] # number of machines
population_size=dfshape[0] # number of jobs
pt=[list(map(int, pt_tmp.iloc[i])) for i in range(population_size)]
ms=[list(map(int,ms_tmp.iloc[i])) for i in range(population_size)]
n=population_size
population=initialization(n,mac)
#samplefitness=calculate(population,n,mac,pt,ms)
#print(samplefitness)
```

```python
INS,SNS,RNS,BNS = [],[],[],[]
ms1=[]
fitness_INS,fitness_SNS,fitness_RNS,fitness_BNS=[],[],[],[]
fac_INS,fac_SNS,fac_RNS,fac_BNS=[],[],[],[]
itr=1
for i in range(itr):
    a=np.random.randint(low=1,high=n+1,size=(n*mac))

currins = ms[0]
currsns = ms[0]
currrns = ms[0]
currbns = ms[0]
b=15000
```

```python
for i in range(b):
    INS.append(currins)
    newins=neighborhoodINS(n,currins)
    currins=newins
    
    SNS.append(currsns)
    newsns=neighborhoodINS(n,currsns)
    currsns=newsns
    
    RNS.append(currrns)
    newrns=neighborhoodINS(n,currrns)
    currrns=newrns
    
    BNS.append(currbns)
    newbns=neighborhoodBNS(n,currbns,a)
    currbns=newbns
    
    
```

```python
for i in range(0,b,n):
    newins=INS[i:i+n]
    newsns=SNS[i:i+n]
    newrns=RNS[i:i+n]
    newbns=BNS[i:i+n]

    fitins=(calculate(population,n,mac,pt,newins))
    fitness_INS.append(fitins)
    fitsns=(calculate(population,n,mac,pt,newsns))
    fitness_SNS.append(fitsns)
    fitrns=(calculate(population,n,mac,pt,newrns))
    fitness_RNS.append(fitrns)
    fitbns=(calculate(population,n,mac,pt,newbns))
    fitness_BNS.append(fitbns)
```

```python
flatten_ins = list(chain.from_iterable(fitness_INS))
flatten_sns = list(chain.from_iterable(fitness_SNS))
flatten_rns = list(chain.from_iterable(fitness_RNS))
flatten_bns = list(chain.from_iterable(fitness_BNS))
```

```python

fac_INS.append(fact(INS))
fac_SNS.append(fact(SNS))
fac_RNS.append(fact(RNS))
fac_BNS.append(fact(BNS))
```
