```python
import numpy as np
import math
import time
import random
import itertools
#import Queue as queue
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
from itertools import permutations
from itertools import chain
import statistics

b=10000
```

```python
with open('epi_ins_abzf.txt') as f:
    epi_INS_abzf = f.read()
    
with open('epi_ins_yn1.txt') as f:
    epi_INS_yn1 = f.read()

with open('epi_ins_la37.txt') as f:
    epi_INS_la37 = f.read()
    
    
epi_ins_abzf,epi_ins_yn1,epi_ins_la37=[],[],[]

temp=0
for ele in epi_INS_abzf:
    
    if (ele=='.'):
        epi_ins_abzf.append(temp)
    elif ele.isdigit():
        temp=temp*10+int(ele)
    elif(ele==' '):
        temp=0
        
temp=0
for ele in epi_INS_yn1:
    
    if (ele=='.'):
        epi_ins_yn1.append(temp)
    elif ele.isdigit():
        temp=temp*10+int(ele)
    elif(ele==' '):
        temp=0
        
temp=0
for ele in epi_INS_la37:
    
    if (ele=='.'):
        epi_ins_la37.append(temp)
    elif ele.isdigit():
        temp=temp*10+int(ele)
    elif(ele==' '):
        temp=0
        
        

    
    
    
    
fig = plt.gcf()
fig.set_size_inches(30.5, 15.5)
indexes = np.arange(1,101)

#epi_ins_abzf=epi_ins_abzf[0:100]
#epi_ins_yn1=epi_ins_yn1[0:100]
#epi_ins_la37=epi_ins_la37[0:100]





plt.clf()
plt.title("Analysis of epistasis value with INS of differennt dataset",fontsize=40)
plt.ylabel("INS value",fontsize=40)
plt.xlabel("Fitness Values",fontsize=40)
plt.plot(indexes, epi_ins_abzf, 'go--', alpha=0.6, linestyle="-", linewidth=1, markersize=8, label="Dataset 1", color="red")
plt.plot(indexes, epi_ins_yn1, 'go--', alpha=0.7, linestyle="--", linewidth=2, markersize=8, label="Dataset 2",color="green")
plt.plot(indexes, epi_ins_la37, 'go--', alpha=0.8, linestyle="-.", linewidth=3, markersize=8, label="Dataset 3",color="blue")

plt.legend(loc="lower right")
plt.savefig("ins_compare.png")
plt.show()
```
