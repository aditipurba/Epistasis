```python
with open('fac_INS.txt') as f:
    fac_INS = f.read()
with open('fac_SNS.txt') as f:
    fac_SNS = f.read()
with open('fac_RNS.txt') as f:
    fac_RNS = f.read()
with open('fac_BNS.txt') as f:
    fac_BNS = f.read()
    
```

```python
def DecimalToBinary(num):
    num=int(num)
    firstZ=0
    bin1=0
    realbin=0
    if num > 1:
        while(num):
            if(bin1==0 and num%2==0):
                firstZ+=1
            bin1=bin1*10+(num%2)
            num=num//2
        while(bin1):
            realbin=realbin*10+(bin1%10)
            bin1=bin1//10
        while(firstZ):
            realbin=realbin*10;
            firstZ-=1;
        return realbin
    else:
        return num

    
    
import math
arr=[]
arr.append(2) 
for i in range (1,512):
    temp=int(math.pow(2,(i-1)))
    #print(temp)
    arr.append( arr[i-1] *2 + temp)
#print(arr)


def epi(num):
    s=str(num)
    nlen=len(s)
    ct0 = 0
    ct1 =  0
    for i in range(nlen):
        if s[i] == '0':
            ct0 = ct0 + 1
        elif s[i] == '1':
            ct1 =ct1 + 1
    divavg=math.pow(2,nlen-1)
    sum1=nlen*(arr[nlen-1]) + divavg*ct1
    ans0=sum1/divavg
    ws = ans0 - (nlen-1)*(arr[nlen]/math.pow(2,nlen))
    xs = nlen +1
    ds = xs - ws
    return ds
   
```

```python
bin_fac_INS,bin_fac_BNS,bin_fac_RNS,bin_fac_SNS=[],[],[],[]
#computing binary value for fac_INS
temp=0
for ele in fac_INS:
    
    if ele.isdigit():
        temp=temp*10+int(ele)
    elif(ele==' '):
        bin_fac_INS.append(DecimalToBinary(int(temp)))
        temp=0

        
#computing binary value for fac_SNS       
temp=0
for ele in fac_SNS:
    
    if ele.isdigit():
        temp=temp*10+int(ele)
    elif(ele==' '):
        bin_fac_SNS.append(DecimalToBinary(int(temp)))
        temp=0
        
        
#computing binary value for fac_RNS       
temp=0
for ele in fac_RNS:
    
    if ele.isdigit():
        temp=temp*10+int(ele)
    elif(ele==' '):
        bin_fac_RNS.append(DecimalToBinary(int(temp)))
        temp=0
        
        
#computing binary value for fac_BNS       
temp=0
for ele in fac_BNS:
    
    if ele.isdigit():
        temp=temp*10+int(ele)
    elif(ele==' '):
        bin_fac_BNS.append(DecimalToBinary(int(temp)))
        temp=0
        
    
#print(bin_fac_INS)  
```

```python
epi_INS,epi_RNS,epi_BNS,epi_SNS=[],[],[],[]
for ele in bin_fac_INS:
    epi_INS.append(epi(ele))

for ele in bin_fac_SNS:
    epi_SNS.append(epi(ele))

for ele in bin_fac_RNS:
    epi_RNS.append(epi(ele))

for ele in bin_fac_BNS:
    epi_BNS.append(epi(ele))
```

```python
#saving epistasis value of fac_INS to file
with open("epi_INS.txt", "w") as fp:  
    for ele in epi_INS:
        fp.write('%s ' % ele)
#saving epistasis value of fac_SNS to file
with open("epi_SNS.txt", "w") as fp:  
    for ele in epi_SNS:
        fp.write('%s ' % ele)
#saving epistasis value of fac_RNS to file
with open("epi_RNS.txt", "w") as fp:  
    for ele in epi_RNS:
        fp.write('%s ' % ele)
#saving epistasis value of fac_BNS to file
with open("epi_BNS.txt", "w") as fp:  
    for ele in epi_BNS:
        fp.write('%s ' % ele)
```
