```python
def DecimalToBinary(num):
    firstZ=0
    bin1=0
    realbin=0
    if num > 1:
        while(num):
            if(bin1==0 and num%2==0):
                firstZ+=1
            bin1=bin1*10+(num%2)
            num=num//2
        while(bin1):
            realbin=realbin*10+(bin1%10)
            bin1=bin1//10
        while(firstZ):
            realbin=realbin*10;
            firstZ-=1;
        return realbin
    else:
        return num

    
    
import math
arr=[]
arr.append(2) 
for i in range (1,512):
    temp=int(math.pow(2,(i-1)))
    #print(temp)
    arr.append( arr[i-1] *2 + temp)
#print(arr)


def epi(num):
    s=str(num)
    nlen=len(s)
    ct0 = 0
    ct1 =  0
    for i in range(nlen):
        if s[i] == '0':
            ct0 = ct0 + 1
        elif s[i] == '1':
            ct1 =ct1 + 1
    divavg=math.pow(2,nlen-1)
    sum1=nlen*(arr[nlen-1]) + divavg*ct1
    ans0=sum1/divavg
    ws = ans0 - (nlen-1)*(arr[nlen]/math.pow(2,nlen))
    xs = nlen +1
    ds = xs - ws
    return ds
   
```
